> Na nossa análise, ao carregar um arquivo CSV no Python, encontramos todos os dados tipificados como `object` devido à natureza do formato CSV, que não especifica os tipos de dados. Isso requer uma inspeção cuidadosa e, quando necessário, uma conversão para os tipos de dados estatísticos apropriados.

Com base na natureza estatística dos dados, aqui está como poderíamos categorizar as colunas do dataset `amazon.csv`:
- `product_id`: Variável nominal, pois é um identificador que categoriza cada produto de forma única.
- `product_name`: Também é uma variável nominal, usada para categorizar produtos por nomes.
- `category`: Variável nominal que agrupa produtos em diferentes categorias.
- `discounted_price` e `actual_price`: Variáveis racionais (contínuas), pois representam quantidades que podem assumir uma gama infinita de valores dentro de um intervalo.
- `discount_percentage`: Uma variável racional (contínua) que indica a proporção do desconto em relação ao preço total.
- `rating`: Uma variável discreta, e representa um ranking em uma escala com ordem (1-5).
- `rating_count`: Uma variável discreta (quantitativa) que conta o número de avaliações.
- `about_product`, `review_title` e `review_content`: Variáveis textuais, podem ser consideradas nominais para a categorização do conteúdo textual.
- `user_id` e `user_name`: Variáveis nominais devido à sua natureza de identificação.
- `review_id`: Variável nominal que serve como um código único para cada avaliação.
- `img_link` e `product_link`: Variáveis nominais que fornecem referências a URLs.

Ao converter para os tipos estatísticos apropriados, podemos aplicar técnicas de análise descritiva, inferencial e preditiva mais precisas, cada uma alinhada com as características dos dados em questão.


```python
# Garantir que as colunas sejam tratadas como strings e tratar valores nulos
amazon_df['discounted_price'] = amazon_df['discounted_price'].fillna('').astype(str)
amazon_df['actual_price'] = amazon_df['actual_price'].fillna('').astype(str)
amazon_df['discount_percentage'] = amazon_df['discount_percentage'].fillna('').astype(str)
amazon_df['rating_count'] = amazon_df['rating_count'].fillna('').astype(str)

# Remover caracteres de moeda e converter para float
amazon_df['discounted_price'] = amazon_df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
amazon_df['actual_price'] = amazon_df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)

# Remover o símbolo de porcentagem e converter para float
amazon_df['discount_percentage'] = amazon_df['discount_percentage'].str.replace('%', '').astype(float) / 100

# Substituir vírgulas em números e converter para int
# A presença de valores nulos ou strings vazias pode causar erros na conversão para int, 
# então vamos tratar isso com uma função lambda que retorna np.nan para strings vazias
amazon_df['rating_count'] = amazon_df['rating_count'].str.replace(',', '').apply(lambda x: np.nan if x == '' else int(x))
```

```python
# Garantir que a coluna 'rating' seja tratada como string
amazon_df['rating'] = amazon_df['rating'].fillna('').astype(str)

# Remover quaisquer caracteres não-numéricos que não sejam o ponto decimal
amazon_df['rating'] = amazon_df['rating'].str.replace('[^0-9.]', '', regex=True)

# Converter strings vazias para NaN e, em seguida, converter para float
amazon_df['rating'] = amazon_df['rating'].apply(lambda x: np.nan if x == '' else float(x))

```

```python
# Agora o rating está certo
print(amazon_df['rating'])
```

#### 3. Quais são as medidas de posição e dispersão?

```python
# 3. Quais são as medidas de posição e dispersão?
# Olhando as estatísticas descritivas para colunas numéricas.
print(amazon_df.describe())

# Para colunas categóricas, vamos olhas a moda das colunas
for col in amazon_df.select_dtypes(include='object').columns:
    print(f'{col} mais frequente: {amazon_df[col].mode()[0]}')
```

#### 4. Como é a distribuição e frequência dos dados?

```python
# 4. Como é a distribuição e frequência dos dados?

# Para variáveis categóricas, vamos gerar um histograma
print(amazon_df['category'].value_counts())

# Para variáveis numéricas, você pode usar visualização, como histogramas:
plt.hist(amazon_df['discounted_price'], bins=50)
plt.title('Distribuição do Preço com Desconto')
plt.xlabel('Preço com Desconto')
plt.ylabel('Frequência')
plt.show()
```

#### 5. Existem correlações significativas entre as variáveis?

```python
# 5. Existem correlações significativas entre as variáveis?

# Seleciona apenas as colunas numéricas para a correlação
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numeric_df = amazon_df.select_dtypes(include=numerics)

# Calcula a matriz de correlação apenas para as colunas numéricas
corr = numeric_df.corr()

# Gera o mapa de calor
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Mapa de Calor das Correlações')
plt.show()
```

Observando o gráfico, entendemos que essa correlação alta entre o preço com desconto e o preço real, de 0.96, mostra que quanto mais caro é o produto, mais alto é o desconto em valor absoluto. É aquela história, quanto mais você gasta, mais você "economiza", né?

Já a relação do percentual de desconto com o preço: é negativa. Ou seja, não espere um descontão percentual nos itens mais caros – a loja não está dando aquele abate tão grande em porcentagem quanto nos produtos mais em conta.

Agora, as notas dos produtos e a quantidade de avaliações não estão dando uma de melhores amigos com o preço ou o desconto. Não é porque o produto é mais caro ou tem um super desconto que ele vai ter notas altas ou um montão de comentários.

E tem mais um detalhe: tem uma leve tendência de produtos mais caros levarem notas um pouco melhores, o que pode ser um indício de que preço mais alto pode ser sinal de qualidade, ou pelo menos de percepção de qualidade.

#### 7. Existem anomalias ou outliers?

```python
# 7. Existem anomalias ou outliers?

# Vamos reutilizar o numerics_df que geramos anteriormente
# Configurando o tamanho da figura para acomodar os subplots
plt.figure(figsize=(15, 10))

# Criando um boxplot para cada coluna numérica
for index, column in enumerate(numeric_df.columns):
    plt.subplot(len(numeric_df.columns), 1, index + 1)
    sns.boxplot(x=numeric_df[column])
    plt.title(f'Boxplot de {column}')
    plt.xlabel(column)
    plt.ylabel('Valores')

# Ajusta o layout para evitar sobreposição
plt.tight_layout()
plt.show()
```

Nos preços com desconto (discounted_price) e nos preços reais (actual_price), tem uma quantidade bem grande de outliers para cima. Isso pode indicar que a maioria dos produtos tem um preço mais em conta, mas tem uns poucos que são bem mais caros. Talvez sejam produtos de categoria premium ou algo assim.

Já na porcentagem de desconto (discount_percentage), o boxplot está bem compacto e não mostra outliers. Parece que os descontos são bem consistentes, mas tem uns poucos produtos com desconto quase zero, que estão ali no limite superior do bigode do boxplot.

O rating tá mais ou menos como a gente esperaria para um e-commerce: a maioria dos produtos tem notas entre 3 e 5, com poucos produtos tendo notas super baixas. Os outliers estão acima de 4.5, o que sugere que tem produtos que são muito bem avaliados pela galera.

Agora, em relação ao rating_count, temos muitos outliers. Isso mostra que tem uns produtos que são muito mais avaliados que a média. Pode ser que esses produtos sejam muito populares ou estão há mais tempo no mercado.

> Lembrando que isso é em relação a base de dados que temos para análise e não necessariamente representa o comportamento no ecommerce inteiro da amazon.
